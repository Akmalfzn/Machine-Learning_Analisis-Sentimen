**TOKENIZING**

```python
nltk.download('punkt')

import string 
import re #regex library


# import word_tokenize & FreqDist from NLTK
from nltk.tokenize import word_tokenize 
from nltk.probability import FreqDist


# ------ Tokenizing ---------

def remove_tweet_special(text):
    # remove tab, new line, ans back slice
    text = text.replace('\\t'," ").replace('\\n'," ").replace('\\u'," ").replace('\\',"")
    # remove non ASCII (emoticon, chinese word, .etc)
    text = text.encode('ascii', 'replace').decode('ascii')
    # remove mention, link, hashtag
    text = ' '.join(re.sub("([@#][A-Za-z0-9]+)|(\w+:\/\/\S+)"," ", text).split())
    # remove incomplete URL
    return text.replace("http://", " ").replace("https://", " ")
                
# ---------------------------------------------------
positive_trainset['tweet'] = positive_trainset['tweet'].apply(remove_tweet_special)
negative_trainset['tweet'] = negative_trainset['tweet'].apply(remove_tweet_special)
# ---------------------------------------------------
positive_testset['tweet'] = positive_testset['tweet'].apply(remove_tweet_special)
negative_testset['tweet'] = negative_testset['tweet'].apply(remove_tweet_special)

#remove number
def remove_number(text):
    return  re.sub(r"\d+", "", text)

# ---------------------------------------------------
positive_trainset['tweet'] = positive_trainset['tweet'].apply(remove_number)
negative_trainset['tweet'] = negative_trainset['tweet'].apply(remove_number)
# ---------------------------------------------------
positive_testset['tweet'] = positive_testset['tweet'].apply(remove_number)
negative_testset['tweet'] = negative_testset['tweet'].apply(remove_number)

#remove punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans("","",string.punctuation))

# ---------------------------------------------------
positive_trainset['tweet'] = positive_trainset['tweet'].apply(remove_punctuation)
negative_trainset['tweet'] = negative_trainset['tweet'].apply(remove_punctuation)
# ---------------------------------------------------
positive_testset['tweet'] = positive_testset['tweet'].apply(remove_punctuation)
negative_testset['tweet'] = negative_testset['tweet'].apply(remove_punctuation)

#remove whitespace leading & trailing
def remove_whitespace_LT(text):
    return text.strip()

# ---------------------------------------------------
positive_trainset['tweet'] = positive_trainset['tweet'].apply(remove_whitespace_LT)
negative_trainset['tweet'] = negative_trainset['tweet'].apply(remove_whitespace_LT)
# ---------------------------------------------------
positive_testset['tweet'] = positive_testset['tweet'].apply(remove_whitespace_LT)
negative_testset['tweet'] = negative_testset['tweet'].apply(remove_whitespace_LT)

#remove multiple whitespace into single whitespace
def remove_whitespace_multiple(text):
    return re.sub('\s+',' ',text)

# ---------------------------------------------------
positive_trainset['tweet'] = positive_trainset['tweet'].apply(remove_whitespace_multiple)
negative_trainset['tweet'] = negative_trainset['tweet'].apply(remove_whitespace_multiple)
# ---------------------------------------------------
positive_testset['tweet'] = positive_testset['tweet'].apply(remove_whitespace_multiple)
negative_testset['tweet'] = negative_testset['tweet'].apply(remove_whitespace_multiple)

# remove single char
def remove_singl_char(text):
    return re.sub(r"\b[a-zA-Z]\b", "", text)

# ---------------------------------------------------
positive_trainset['tweet'] = positive_trainset['tweet'].apply(remove_singl_char)
negative_trainset['tweet'] = negative_trainset['tweet'].apply(remove_singl_char)
# ---------------------------------------------------
positive_testset['tweet'] = positive_testset['tweet'].apply(remove_singl_char)
negative_testset['tweet'] = negative_testset['tweet'].apply(remove_singl_char)

# NLTK word rokenize 
def word_tokenize_wrapper(text):
    return word_tokenize(text)

# ---------------------------------------------------
positive_trainset['tweet_tokens'] = positive_trainset['tweet'].apply(word_tokenize_wrapper)
negative_trainset['tweet_tokens'] = negative_trainset['tweet'].apply(word_tokenize_wrapper)
# ---------------------------------------------------
positive_testset['tweet_tokens'] = positive_testset['tweet'].apply(word_tokenize_wrapper)
negative_testset['tweet_tokens'] = negative_testset['tweet'].apply(word_tokenize_wrapper)

print(positive_trainset['tweet_tokens'])
print('\n\n\n')
print(negative_trainset['tweet_tokens'])
print('\n\n\n')
print(positive_testset['tweet_tokens'])
print('\n\n\n')
print(negative_testset['tweet_tokens'])
```

**NORMALIZATION/FORMALIZATION**

```python
normalizad_word = pd.read_excel("normalisasi.xlsx")

normalizad_word_dict = {}

for index, row in normalizad_word.iterrows():
    if row[0] not in normalizad_word_dict:
        normalizad_word_dict[row[0]] = row[1] 

def normalized_term(document):
    return [normalizad_word_dict[term] if term in normalizad_word_dict else term for term in document]

# ---------------------------------------------------
positive_trainset['tweet_normalized'] = positive_trainset['tweet_tokens'].apply(normalized_term)
negative_trainset['tweet_normalized'] = negative_trainset['tweet_tokens'].apply(normalized_term)
# ---------------------------------------------------
positive_testset['tweet_normalized'] = positive_testset['tweet_tokens'].apply(normalized_term)
negative_testset['tweet_normalized'] = negative_testset['tweet_tokens'].apply(normalized_term)

print(positive_trainset['tweet_normalized'])
print('\n\n\n')
print(negative_trainset['tweet_normalized'])
print('\n\n\n')
print(positive_testset['tweet_normalized'])
print('\n\n\n')
print(negative_testset['tweet_normalized'])
```

**FILTERING (Stopword Removal) and (Removing 'PPKM' & 'DARURAT')**

```python
nltk.download('stopwords')
from nltk.corpus import stopwords

# ----------------------- get stopword from NLTK stopword -------------------------------
# get stopword indonesia
list_stopwords = stopwords.words('indonesian')


# ---------------------------- manualy add stopword  ------------------------------------
# append additional stopword (MENGHAPUS 'PPKM' & 'DARURAT')
list_stopwords.extend(["yg", "dg", "rt", "dgn", "ny", "d", 'klo', 
                       'ppkm', 'darurat', 'kalo', 'amp', 'biar', 'bikin', 'bilang', 'birthday', 'doang', 'dong', 'cari', 'itu',
                       'krn', 'karna', 'nya', 'nih', 'sih', 'udh', 'sampai', 'kenapa', 'duluan', 'ada', 'abis', 'ugh', 'pengen',
                       'si', 'tau', 'tuh', 'utk', 'ya', 'trs', 'sm', 'padahal', 'lagi', 'dpt', 'dapat', 'dapet', 'ken', 'mlu',
                       'jd', 'sdh', 'aja', 'n', 't', 'pas', 'psbb', 'yang', 'apa', 'yang', 'banyak', 'buat', 'pls', 'mulu', 'cari',
                       'nyg', 'hehe', 'pen', 'u', 'nan', 'loh', 'rt', 'emg', 'buat', 'sdg', 'pada', 'pda', 'allah', 'ydh', 'yaudah',
                       '&amp', 'banget', 'yah'])

# ----------------------- add stopword from txt file ------------------------------------
# read txt stopword using pandas
txt_stopword = pd.read_csv("stopwords.txt", names= ["stopwords"], header = None)

# convert stopword string to list & append additional stopword
list_stopwords.extend(txt_stopword["stopwords"][0].split(' '))

# ---------------------------------------------------------------------------------------

# convert list to dictionary
list_stopwords = set(list_stopwords)

# tidak menghilangkan kata negasi
for word in list_stopwords.copy():
    if (word == 'tidak'):
      list_stopwords.remove('tidak')
    elif (word == 'kagak'):
      list_stopwords.remove('kagak')
    elif (word == 'tak'):
      list_stopwords.remove('tak')
    elif (word == 'enggak'):
      list_stopwords.remove('enggak')
    elif (word == 'bukan'):
      list_stopwords.remove('bukan')
    elif (word == 'masalah'):
      list_stopwords.remove('masalah')
    elif (word == 'guna'):
      list_stopwords.remove('guna')

#remove stopword pada list token
def stopwords_removal(words):
    return [word for word in words if word not in list_stopwords]

# ---------------------------------------------------
positive_trainset['tweet_tokens_WSW'] = positive_trainset['tweet_normalized'].apply(stopwords_removal)
negative_trainset['tweet_tokens_WSW'] = negative_trainset['tweet_normalized'].apply(stopwords_removal)
# ---------------------------------------------------
positive_testset['tweet_tokens_WSW'] = positive_testset['tweet_normalized'].apply(stopwords_removal)
negative_testset['tweet_tokens_WSW'] = negative_testset['tweet_normalized'].apply(stopwords_removal)

print(positive_trainset['tweet_tokens_WSW'])
print('\n\n\n')
print(negative_trainset['tweet_tokens_WSW'])
print('\n\n\n')
print(positive_testset['tweet_tokens_WSW'])
print('\n\n\n')
print(negative_testset['tweet_tokens_WSW'])
```
